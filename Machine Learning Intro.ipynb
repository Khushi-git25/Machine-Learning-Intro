***Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).***

Answer:

---

### **1. Artificial Intelligence (AI):**

* **Definition:** AI is the broadest concept—it refers to the science of creating systems that can perform tasks that normally require human intelligence.
* **Goal:** Make machines *think and act like humans* (to some extent).
* **Examples:**

  * Virtual assistants (Siri, Alexa)
  * Game-playing agents (like AlphaGo)
  * Chatbots

---

### **2. Machine Learning (ML):**

* **Definition:** ML is a *subset of AI* that focuses on creating systems that learn patterns from data and improve performance without being explicitly programmed.
* **Goal:** Enable machines to *learn from experience/data*.
* **Examples:**

  * Spam email detection
  * Predicting house prices
  * Recommendation systems (Netflix, Amazon)

---

### **3. Deep Learning (DL):**

* **Definition:** DL is a *subset of ML* that uses **artificial neural networks with many layers** to model complex patterns in large amounts of data.
* **Goal:** Solve highly complex tasks by mimicking the human brain’s structure.
* **Examples:**

  * Image recognition (e.g., detecting objects in photos)
  * Speech recognition
  * Self-driving cars (lane detection, obstacle avoidance)

---

### **4. Data Science (DS):**

* **Definition:** Data Science is a broader field that combines **statistics, data analysis, machine learning, and domain expertise** to extract knowledge and insights from data.
* **Goal:** *Understand and use data effectively* for decision-making.
* **Examples:**

  * Business analytics and dashboards
  * Customer segmentation
  * Predictive modeling for sales/finance

---

### **Key Differences in Scope:**

| Term   | Scope & Focus                                                          | Relation to Others  |
| ------ | ---------------------------------------------------------------------- | ------------------- |
| **AI** | Broad field of making machines "intelligent"                           | Umbrella concept    |
| **ML** | Subset of AI, focuses on *learning from data*                          | A way to achieve AI |
| **DL** | Subset of ML, uses *neural networks with many layers*                  | A way to do ML      |
| **DS** | Broader than AI/ML, focuses on extracting insights and value from data | Uses ML/DL as tools |

---

 In short:

* **AI** = the goal (make machines smart)
* **ML** = the method (learn from data)
* **DL** = the advanced method (neural networks)
* **DS** = the practice (analyzing and using data for insights)

***Question 2: What are the types of machine learning? Describe each with one
real-world example.***

Answer: Machine Learning (ML) is typically classified into **three main types** (sometimes a fourth is added).

---

## **1. Supervised Learning**

* **Definition:** The model is trained on a *labeled dataset* (input data + correct output). The algorithm learns the mapping between inputs and outputs.
* **Goal:** Predict outcomes for new, unseen data.
* **Techniques:** Regression, Classification.
* **Example:**

  * **Spam Email Detection**

    * Input: Email text (features).
    * Output (label): "Spam" or "Not Spam".
    * The model learns from historical emails with labels and can then classify new ones.

---

## **2. Unsupervised Learning**

* **Definition:** The model is trained on *unlabeled data* (only inputs, no correct outputs). The algorithm tries to find hidden patterns, structures, or groupings in the data.
* **Goal:** Discover insights from unstructured data.
* **Techniques:** Clustering, Dimensionality Reduction.
* **Example:**

  * **Customer Segmentation in Marketing**

    * Input: Customer data (age, income, purchase history).
    * No labels (like "loyal" or "churn").
    * The model groups customers into clusters (e.g., budget buyers, premium shoppers).

---

## **3. Reinforcement Learning (RL)**

* **Definition:** An agent learns by *interacting with an environment*, receiving rewards or penalties for actions, and improving its strategy over time.
* **Goal:** Maximize cumulative rewards through trial and error.
* **Techniques:** Q-learning, Deep Q-Networks.
* **Example:**

  * **Self-Driving Cars**

    * The car (agent) learns to drive by interacting with the environment (road, traffic).
    * Positive reward: staying in lane, avoiding obstacles.
    * Negative reward: collisions or breaking traffic rules.

---

## *(Optional 4th)* Semi-Supervised Learning

* **Definition:** Uses both *labeled and unlabeled data*. A small amount of labeled data guides the learning on a larger set of unlabeled data.
* **Example:**

  * **Medical Diagnosis**

    * Labeled data: A few X-ray images with diagnoses (tumor present/absent).
    * Unlabeled data: Thousands of X-ray images without labels.
    * The algorithm learns effectively by combining both.

---

### **Quick Comparison Table**

| Type                        | Data Used                  | Goal                             | Example               |
| --------------------------- | -------------------------- | -------------------------------- | --------------------- |
| **Supervised Learning**     | Labeled data               | Predict outputs                  | Spam detection        |
| **Unsupervised Learning**   | Unlabeled data             | Find hidden patterns             | Customer segmentation |
| **Reinforcement Learning**  | Rewards & penalties        | Learn best actions over time     | Self-driving cars     |
| **Semi-Supervised** (extra) | Mix of labeled & unlabeled | Improve accuracy with few labels | Medical diagnosis     |

***Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning?***

Answer:

---

## **1. Overfitting**

* **Definition:**

  * Overfitting occurs when a model **learns the training data too well**, including noise and random fluctuations, so it performs **poorly on new/unseen data**.
* **Characteristics:**

  * High accuracy on training data
  * Low accuracy on test/validation data
  * Complex model with too many parameters
* **Example:**

  * A polynomial regression of degree 10 perfectly fits 10 training points but fails to predict new points accurately.

---

## **2. Underfitting**

* **Definition:**

  * Underfitting occurs when a model is **too simple to capture the underlying patterns** in the data.
* **Characteristics:**

  * Low accuracy on training data
  * Low accuracy on test/validation data
  * Model cannot capture relationships (high bias)
* **Example:**

  * A linear regression trying to fit a strongly quadratic relationship.

---

## **3. Bias-Variance Tradeoff**

* **Definition:**

  * The **bias-variance tradeoff** is the balance between:

    * **Bias:** Error due to overly simple assumptions (underfitting)
    * **Variance:** Error due to sensitivity to small fluctuations in training data (overfitting)
* **Goal:** Find the model complexity that **minimizes total error** (sum of bias² + variance + irreducible error).

**Visual Analogy:**

| Model Type    | Bias     | Variance | Fit Description     |
| ------------- | -------- | -------- | ------------------- |
| Simple model  | High     | Low      | Underfits           |
| Complex model | Low      | High     | Overfits            |
| Optimal model | Moderate | Moderate | Best generalization |

---

### **Key Takeaways**

1. **Underfitting** → model too simple → high bias → poor training & test performance.
2. **Overfitting** → model too complex → high variance → excellent training but poor test performance.
3. **Bias-Variance Tradeoff** → balance model complexity to minimize total prediction error.

***Question 4: What are outliers in a dataset, and list three common techniques for handling them?***

Answer:

---

## **What are Outliers?**

* **Definition:** Outliers are data points that **differ significantly** from the majority of the data.
* They may occur due to **errors** (data entry mistakes, measurement issues) or may represent **rare but valid events**.
* **Effect:** Outliers can **skew statistical results**, distort models (like regression), and reduce accuracy.

**Example:**
If most students score between 60–90 on a test, but one student scores **5** or **100**, those are potential outliers.

---

## **Three Common Techniques to Handle Outliers:**

### **1. Removal (Deleting Outliers)**

* Remove extreme values that are clearly due to errors or irrelevant to analysis.
* **Example:** If someone’s age is recorded as **300**, it’s safe to remove.
*  Use with caution: only when outliers are *definitely incorrect*.

---

### **2. Transformation (Reducing Impact)**

* Apply mathematical transformations to **reduce the effect** of outliers.
* **Methods:** Logarithmic, square root, or Box-Cox transformations.
* **Example:** If income data has a few billionaires, applying a log transform makes the distribution less skewed.

---

### **3. Imputation / Capping (Winsorization)**

* Replace extreme values with more reasonable values.
* **Methods:**

  * **Imputation:** Replace with mean, median, or mode.
  * **Capping (Winsorizing):** Replace outliers with a threshold (e.g., 1st and 99th percentile).
* **Example:** If most house prices are under ₹2 crore but one entry is ₹200 crore, cap it at the 99th percentile.

---

 **Summary:**
Outliers = unusual/extreme values.
**Handling methods:**

1. **Removal** (delete them if errors).
2. **Transformation** (reduce effect).
3. **Imputation/Capping** (replace or limit).

---

***Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.***

Answer:

---

## **Process of Handling Missing Values**

1. **Identify Missing Values**

   * Check the dataset for `NaN`, `NULL`, or blanks using functions like `.isnull()` in Python (Pandas).

2. **Understand the Cause**

   * Missing values may occur due to:

     * Data entry errors
     * Sensor/measurement failure
     * Non-response in surveys
     * Irrelevant features

3. **Decide on a Strategy**

   * Based on the percentage and importance of missing data, you can:

     * **Remove** rows/columns (if too many missing values and not critical).
     * **Impute** (fill in missing values with estimated values).

4. **Apply Imputation Techniques**

   * Choose appropriate methods depending on whether data is **numerical** or **categorical**.

---

## **Imputation Techniques**

### **1. For Numerical Data:**

* **Mean/Median Imputation**

  * Replace missing values with the mean or median of the column.
  * **When to use:**

    * **Mean** → for normally distributed data.
    * **Median** → for skewed data (e.g., income, house prices).
  * **Example:**
    If the column `Age = [25, 30, NaN, 40]`,

    * Mean = 31.6 → replace NaN with 31.6.
    * Median = 30 → replace NaN with 30.

---

### **2. For Categorical Data:**

* **Mode Imputation**

  * Replace missing values with the most frequent category (mode).
  * **Example:**
    If `Gender = [Male, Female, NaN, Female]`,

    * Mode = "Female" → replace NaN with "Female".

---

 **Summary:**

* **Process:** Detect → Analyze → Decide → Impute/Remove.
* **Numerical Data Imputation:** Mean/Median.
* **Categorical Data Imputation:** Mode.

---

Question 6: Write a Python program that:

● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.

● Prints the class distribution


```python
from sklearn.datasets import make_classification
import numpy as np
from collections import Counter

X, y = make_classification(n_samples=1000,
                           n_features=10,
                           n_informative=5,
                           n_redundant=2,
                           n_clusters_per_class=1,
                           weights=[0.9, 0.1],
                           random_state=42)

class_distribution = Counter(y)
print("Class distribution:", class_distribution)

for cls, count in class_distribution.items():
    print(f"Class {cls}: {count} samples ({count/len(y)*100:.2f}%)")
```

Question 8: Write a Python script to:

● Generate 1000 samples from a normal distribution.

● Introduce 50 random missing values.

● Fill missing values with the column mean.

● Plot a histogram before and after imputation.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

df = pd.DataFrame(data, columns=['Value'])

missing_indices = np.random.choice(df.index, 50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

df_before = df.copy()

mean_value = df['Value'].mean()
df['Value'].fillna(mean_value, inplace=True)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df_before['Value'].dropna(), bins=30, edgecolor='black', alpha=0.7)
plt.title("Histogram Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(df['Value'], bins=30, edgecolor='black', alpha=0.7, color='orange')
plt.title("Histogram After Imputation (Mean)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:

● Missing ages,

● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),

● Categorical variables like payment method.

Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.

Answer:


---

## **1. Exploratory Data Analysis (EDA)**

* **Understand the dataset**: Check distributions, missing values, outliers, and class imbalance.
* **Visualizations**: Histograms for numerical features, boxplots for detecting outliers, bar plots for categorical distributions.

---

## **2. Handling Missing Data**

* **Missing Ages:**

  * **Option 1:** Impute with **median** (robust to outliers).
  * **Option 2:** Use **KNN or regression imputation** based on other features like transaction amount or customer segment.
* **Categorical missing values:** Fill with **mode** or a special category like “Unknown.”

---

## **3. Handling Outliers**

* **Transaction Amount Outliers:**

  * Detect using **IQR method** or **Z-scores**.
  * **Option 1:** Cap values at the 1st and 99th percentile (Winsorization).
  * **Option 2:** Apply **log transformation** to reduce skewness.
  * Outliers might also be informative for fraud detection; flag them as a separate feature.

---

## **4. Handling Imbalanced Target**

* **Problem:** Fraud vs. non-fraud is highly imbalanced.
* **Solutions:**

  * **Resampling:**

    * **Oversampling** the minority class (e.g., SMOTE).
    * **Undersampling** the majority class.
  * **Class Weights:** Use algorithms that can adjust for imbalance (e.g., `class_weight='balanced'` in scikit-learn).
* **Evaluation metrics:** Use **precision, recall, F1-score, PR-AUC** instead of accuracy.

---

## **5. Encoding Categorical Variables**

* **Payment Method (categorical):**

  * **Low cardinality:** Use **one-hot encoding**.
  * **High cardinality (e.g., store IDs):** Use **target encoding** or **frequency encoding**.
* Ensure **no data leakage** when using target encoding: fit on training data only.

---

## **6. Feature Scaling (if needed)**

* **Tree-based models:** No scaling required.
* **Distance-based models (KNN, SVM):** Use **StandardScaler** or **RobustScaler**.
* For skewed numerical features (transaction amount), consider **log-transform** before scaling.

---

## **7. Feature Engineering**

* Create new features that can help fraud detection:

  * Transaction frequency per customer.
  * Average transaction amount per customer.
  * Time-based features: hour of day, day of week.
  * Flag for large transactions or outliers.

---

## **8. Train-Test Split and Validation**

* **Use stratified split** to maintain the fraud/non-fraud ratio in training and testing sets.
* **Cross-validation:** Use **stratified K-fold** to ensure minority class representation.

---

### **Summary Table**

| Step                 | Method / Technique                                                       |
| -------------------- | ------------------------------------------------------------------------ |
| Missing data         | Median (numeric), Mode or “Unknown” (categorical), KNN/Iterative Imputer |
| Outliers             | IQR/Z-score detection, Capping, Log transform, Flagging                  |
| Imbalanced target    | SMOTE, undersampling, class weights                                      |
| Categorical encoding | One-hot (low cardinality), Target/frequency encoding (high cardinality)  |
| Scaling              | StandardScaler, RobustScaler (if required)                               |
| Feature engineering  | Transaction frequency, average amount, time features, outlier flags      |
| Validation           | Stratified train-test split, stratified K-fold CV                        |

---

This plan ensures **clean, balanced, and properly encoded data**, ready for training a robust ML model for fraud detection.
